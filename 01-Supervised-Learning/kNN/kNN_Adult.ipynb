```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
#Bước 1: Bắt đầu với tập dữ liệu đã được gán nhãn và điểm dữ liệu mới
#Gán nhãn cột
columns = [
    "age", "workclass", "fnlwgt", "education", "education-num",
    "marital-status", "occupation", "relationship", "race", "sex",
    "capital-gain", "capital-loss", "hours-per-week", "native-country", "income"]
#Đọc dữ liệu
df = pd.read_csv('adult.data', names=columns, sep=',', skipinitialspace=True)
#Làm sạch dữ liệu
df.replace('?', np.nan, inplace=True)
df.dropna(inplace=True)
#Dữ liệu đầu vào x và nhãn y
label = LabelEncoder()
y_data = label.fit_transform(df['income'])
X_raw = df.drop(['income', 'education'], axis=1)
#Chia dữ liệu train/Test
X_data = pd.get_dummies(X_raw, drop_first=True).values
X_train, X_test, y_train, y_test = train_test_split(
    X_data, y_data, test_size=0.2, random_state=42)
#Bước 2: Tính khoảng cách từ điểm mới đến các điểm huấn luyện
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
#Bước 3: Tìm k láng giềng gần nhất
knn_classifier = KNeighborsClassifier(n_neighbors=3)
#Bước 4: Đưa ra dự đoán
knn_classifier.fit(X_train, y_train)
y_pred = knn_classifier.predict(X_test)
print("Predicted results using kNN:", accuracy_score(y_test, y_pred))
```
